```python
# from google.colab import drive
# drive.mount('/content/drive/')
path = "/content/drive/Shareddrives/KAIST FELab Offline Team/Task/Data/"
import pandas as pd
import numpy as np
import copy
# df1 = pd.read_csv(path + "coin_mv2.csv")
df1 = pd.read_csv('coin_mv2.csv',index_col='Date',parse_dates=True)
df1 = df1[::-1]
df1
```

```python
# df2 = pd.read_csv(path + "coin_price.csv")
df2 = pd.read_csv('coin_price.csv',index_col='Date',parse_dates=True)
df2 = df2[::-1]
df2
```

```python
df3 = pd.read_csv('코인모음.csv',index_col='category',parse_dates=True)
df3_ind = df3.index
lst_coin = []
# df4 = df1.copy()
for i in df3_ind:
    b = df3.loc[i]
    a= b.tolist()
    c= a[0].split(',')
    lst_coin.append(c)
lst_coin
```

```python
price_mv = []
price_pr = []
for i in range(len(lst_coin)):
    df4 = df1.copy()
    df5 = df2.copy()
    df6 = df1.copy()
    df7 = df2.copy()
    for j in df1.columns:
        if j not in lst_coin[i]:
            df4 = df4.drop(j, axis=1)
            df6 = df6.drop(j, axis=1)
            df5 = df5.drop(j, axis=1)
            df7 = df7.drop(j, axis=1)
    df4['total_mv'] = df4.sum(axis=1)
    df5['total_pr'] = df5.sum(axis=1)
    for k in df6.columns:
        df4[k] = df4[k]/df4['total_mv']
        df5[k] = df5[k]/df5['total_pr']
    df4 = df4.drop('total_mv', axis=1)
    df5 = df5.drop('total_pr', axis=1)
    df8 = df4*df7
    df9 = df5*df7
    df10 = df8.sum(axis=1)
    df11 = df9.sum(axis=1)
    price_mv.append(df10)
    price_pr.append(df11)
for i in range(len(price_mv)):
    if i == 0:
        df_mv = price_mv[0]
    else:
        df_mv = pd.concat([df_mv,price_mv[i]], axis=1)
for j in range(len(price_pr)):
    if j == 0:
        df_pr = price_pr[0]
    else:
        df_pr = pd.concat([df_pr,price_pr[j]], axis=1)
df_mv.columns = ['price_mv_boostvc_portfolio', 'price_mv_dcg_portfolio', 'price_mv_electric_capital_portfolio', 'price_mv_huobi_capital_portfolio','price_mv_1confirmation_portfolio', 'price_mv_placeholder_ventures_portfolio']
df_pr.columns = ['price_pr_boostvc_portfolio', 'price_pr_dcg_portfolio', 'price_pr_electric_capital_portfolio', 'price_pr_huobi_capital_portfolio','price_pr_1confirmation_portfolio', 'price_pr_placeholder_ventures_portfolio']
df_pr
```

```python
price_dj = []
for i in range(len(lst_coin)):
    df13 = df2.copy()
    for j in df1.columns:
        if j not in lst_coin[i]:
            df13 = df13.drop(j,axis=1)
    df14 = df13/len(df13.columns)
    df15 = df13.sum(axis=1)
    price_dj.append(df15)
    
for i in range(len(price_dj)):
    if i == 0:
        df_dj = price_dj[0]
    else:
        df_dj = pd.concat([df_dj,price_dj[i]], axis=1)

df_dj.columns = ['price_dj_boostvc_portfolio', 'price_dj_dcg_portfolio', 'price_dj_electric_capital_portfolio', 'price_dj_huobi_capital_portfolio','price_dj_1confirmation_portfolio', 'price_dj_placeholder_ventures_portfolio']
df_dj
```

```python
df_mv.to_csv('price_mv.csv')
df_pr.to_csv('price_pr.csv')
df_dj.to_csv('price_dj.csv')
```
