```python
from decimal import Decimal, ROUND_HALF_UP
import numpy as np


def absorption_ratio(
    asset_returns: np.ndarray, fraction_eigenvectors
) -> float:
    n_assets, _ = asset_returns.shape
    cov = np.cov(asset_returns)
    eig = np.linalg.eigvals(cov)
    eig_sorted = sorted(eig)
    num_eigenvalues = int(
        Decimal(fraction_eigenvectors * n_assets).to_integral_value(
            rounding=ROUND_HALF_UP
        )
    )
    return sum(eig_sorted[len(eig_sorted) - num_eigenvalues :]) / np.trace(cov)
```

```python
port_ret = []
port_risk = []
port_weight = []
sharpe_ratio  = []
        
daily_ret = return_data
annual_ret = daily_ret.mean()*757 # *이후에 대상기간 곱해주어야 함
daily_cov = daily_ret.cov()
annual_cov = daily_cov * 757 # *이후에 대상기간 곱해주어야 함

def monte_carlo_sim(stock_list):
    for _ in range(20000):
        weights = np.random.random(len(stock_list))
        total_weight = np.sum(weights)
        weights /= total_weight

        returns = np.dot(weights, annual_ret)
        risk = np.sqrt(np.dot(weights.T, np.dot(annual_cov, weights)))

        port_ret.append(returns)
        port_risk.append(risk)
        port_weight.append(weights)
        sharpe_ratio.append(returns/risk)

    portfolio = {'Returns' : port_ret, 'Risk' : port_risk, 'Sharpe' : sharpe_ratio}
    for i, s in enumerate(stock_list):
        portfolio[s] = [weight[i] for weight in port_weight]

    portfolio = pd.DataFrame(portfolio)
    max_sharpe = portfolio.loc[portfolio['Sharpe'] == portfolio['Sharpe'].max()]
    min_risk = portfolio.loc[portfolio['Risk'] == portfolio['Risk'].min()]

    # portfolio.plot.scatter(x='Risk', y='Returns',grid=True)
    # plt.title("Efficient Frontier")
    # plt.xlabel('Risk')
    # plt.ylabel('Expected Returns')

    portfolio.plot.scatter(x='Risk', y='Returns', c='Sharpe', cmap = 'viridis',edgecolors='k', grid=True)
    plt.scatter(x=min_risk['Risk'], y=min_risk['Returns'], c='r', marker='X', s=200)
    plt.scatter(x=max_sharpe['Risk'], y=max_sharpe['Returns'], c='r', marker='*', s=300)
    plt.title("Efficient Frontier")
    plt.xlabel('Risk')
    plt.ylabel('Expected Returns')
    plt.show()
    print("max sharpe : ", max_sharpe[['Returns', 'Risk', 'Sharpe']])
    print("min_risk:", min_risk[['Returns', 'Risk', 'Sharpe']])

    return portfolio
```

```python
monte_carlo_sim(stock_list)
```

2) 시스템적 리스크 관점 : 
<br>
<br>
CoVaR(Comovement Value at Risk)가 여타의 cluster들에 비해 0과 가깝게 측정되었음. 이는 분석 대상으로 사용한 암호화폐 cluster가 기존 나스닥에 편입되어있던 자산들에 비해 추종성이 낮은 것으로 볼 수 있음. 따라서 주식 시장에서 암호화폐 ETF가 새로운 자산으로서의 역할을 할 수 있을 것으로 사료됨
<br>
<br>
또한 암호화폐 ETF 편입 전/후를 기준으로 Absorption Ratio를 측정해본 결과, 암호화폐 etf를 편입한 시장은 암호화폐 etf를 편입하지 않은 시장에 비해 더 낮은 Absorption Ratio를 보임으로써(n=2일 경우 0.74, 0.77), 암호화폐 etf의 편입은 시스템적 리스크 관점에서도 긍정적인 결과를 낳을 수 있음을 보임
<br>
<br>
3) Efficient Frontier & Sharpe Ratio : 
<br>
<br>
효율적 투자선을 그려본 결과, 기존 나스닥 시장의 max_sharpe일 경우의 sharpe ratio는 1.09, min_risk일 경우의 sharpe ratio는 0.25로 측정된 반면 암호화폐 etf를 포함한 시장의 max sharpe의 sharpe ratio는 1.17, min_risk일 경우의 sharpe ratio는 0.66으로 측정되어 암호화폐 etf를 편입할 경우 더 높은 sharpe ratio를 얻을 수 있음을 보임

결론 : 기존 시장의 암호화폐 etf편입은 시스템 리스크 및 포트폴리오 관점에서 새로운 자산의 역할을 할 수 있을 뿐만 아니라 수익적 측면에서도 긍정적 결과를 낳을 것으로 사료됨
