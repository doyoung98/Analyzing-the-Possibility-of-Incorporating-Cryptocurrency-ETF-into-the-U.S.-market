```python
import datetime
import pandas as pd
import numpy as np
```

```python
price_us = pd.read_csv("etf_close.csv", index_col=0)
price_mv = pd.read_csv("price_mv.csv", index_col=0)
price_pr = pd.read_csv("price_pr.csv", index_col=0)
price_dj = pd.read_csv("price_dj.csv", index_col=0)
```

```python
price_us.dropna(axis=1, inplace=True)
price_raw1 = pd.concat([price_us, price_mv], axis=1 )
price_raw1.dropna(axis=0, inplace=True)
# price_raw1 = price_raw1.iloc[1:]

price_raw1
# price_us
# price_mv
# price_raw
```

```python
price_raw2 = pd.concat([price_us, price_pr], axis=1 )
price_raw2.dropna(axis=0, inplace=True)
# price_raw2 = price_raw2.iloc[1:]

price_raw2
```

```python
price_raw3 = pd.concat([price_us, price_dj], axis=1 )
price_raw3.dropna(axis=0, inplace=True)
# price_raw3 = price_raw3.iloc[1:]

price_raw3
```

```python
raw_return_mv_df = price_raw1.pct_change()
raw_return_mv_df = raw_return_mv_df.iloc[1:, :]
raw_return_mv_df
```

```python
raw_return_pr_df = price_raw2.pct_change()
raw_return_pr_df = raw_return_pr_df.iloc[1:, :]
raw_return_pr_df
```

```python
raw_return_dj_df = price_raw3.pct_change()
raw_return_dj_df = raw_return_dj_df.iloc[1:, :]
raw_return_dj_df
```

```python
log_return_mv_df = np.log(price_raw1/price_raw1.shift(1))
log_return_mv_df = log_return_mv_df.iloc[1:, :]
log_return_mv_df
```

```python
log_return_pr_df = np.log(price_raw2/price_raw2.shift(1))
log_return_pr_df = log_return_pr_df.iloc[1:, :]
log_return_pr_df
```

```python
log_return_dj_df = np.log(price_raw3/price_raw3.shift(1))
log_return_dj_df = log_return_dj_df.iloc[1:, :]
log_return_dj_df
```

```python
raw_return_dj_df.to_csv("raw_dj2.csv")
raw_return_mv_df.to_csv("raw_mv2.csv")
raw_return_pr_df.to_csv("raw_pr2.csv")

log_return_dj_df.to_csv("log_dj2.csv")
log_return_mv_df.to_csv("log_mv2.csv")
log_return_pr_df.to_csv("log_pr2.csv")
```
