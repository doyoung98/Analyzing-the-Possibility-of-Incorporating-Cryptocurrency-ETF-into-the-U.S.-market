```python
import numpy as np
import sklearn
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import fcluster
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet
from scipy.stats import norm
import time
```

```python
# kmeans clustering이 완료된 data받기
data = pd.read_csv('../data/kmeans.csv', index_col = 0)
data
```

```python
data_cluster_list = []
data1 = data.transpose()
predict_value_counts = len(data1['cluster'].value_counts())

for i in range(predict_value_counts):
    src = data1[data1['cluster'] == i]
    src = src.drop(['cluster'], axis=1)
    src = src.transpose()
    data_cluster_list.append(src)
```

```python
#클러스터별 수익률을 데이터프레임 형태로 구성

return_lst = []
col_name=[]
for i in range(predict_value_counts):
    df = data_cluster_list[i]
#     df = df.transpose()
#     df = df.drop(['cluster'])
    return_df = df.mean(axis=1)
    return_lst.append(return_df)
    str_name = 'cluster '+str(i)
    col_name.append(str_name)

for i in range(len(return_lst)):
    if i == 0:
        df_return = return_lst[i]
    else:
        df_return = pd.concat([df_return,return_lst[i]], axis=1)
df_return.columns = col_name
# df_return.to_csv('df_return.csv', index = False)
df_return
```

```python
# 1개의 클러스터와 전체 시스템의 수익률간 상관관계의 모음(120개)
# 200일씩 rolling

df_return['total_cluster'] = df_return.mean(axis=1)
rho_lst = []
for i in range(557):
    df= df_return.iloc[i:i+200,:]
    rho_lst2 = []
    for j in range(len(df.columns)-1):
        df1 = df[[df.columns[j],'total_cluster']]
        rho = df1.corr().iloc[0,1]
        rho_lst2.append(rho)
    rho_lst.append(rho_lst2)
len(rho_lst)
```

```python
#전체 시스템의 수익률 표준편차
# 200일씩 rolling

sigma_lst = []
fin = []
df_return2 = df_return.iloc[:,0:120]
df_return2
for i in range(557):
    df2 = df_return2.iloc[i:i+200,:]
    corr = df2.corr()
    sum_ax1 = corr.sum(axis=1)
    sum_fin = sum_ax1.sum(axis=0)
    fin.append(sum_fin)
    sum_fin = (sum_fin/((len(df2.columns))**2))**0.5
    sigma_lst.append(sum_fin)
```

```python
# 수식에 의하면, 한 군집과 전체 시스템의 CoVaR의 변화량은 rho*sigma(system)에 비례함
# rho와 sigma(system)의 곱을 구하는 과정
covar_lst = []
for i in range(len(rho_lst)):
    list1= rho_lst[i]
    sigma =sigma_lst[i]
    list2 = []
    for j in list1:
        list2.append(j*sigma)
    covar_lst.append(list2)
```

```python
# rolling한 CoVaR을 plot하는 과정
import matplotlib.pyplot as plt
for i in range(len(covar_lst)):
    date = df_return.index[i+200]
    date_lst = [date]*120
    plt.plot(date_lst,covar_lst[i])
plt.show()
```
