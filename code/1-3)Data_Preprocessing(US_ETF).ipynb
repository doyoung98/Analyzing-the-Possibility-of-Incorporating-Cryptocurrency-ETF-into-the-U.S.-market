### Investing.com 으로부터 US ETF download

```python
etf_dict_lst = investpy.etfs.get_etfs_dict(country='united states')
etf_ticker_lst = []

for etf_dict in etf_dict_lst:
  ticker = etf_dict['symbol']
  etf_ticker_lst.append(ticker)

close_df = pd.DataFrame(etf_ticker_lst)
close_df = close_df.transpose()
close_df 
```

```python
print(len(close_df.columns))
```

### ETF Close df

```python
start = datetime.datetime(2018,12,31)
# end = datetime.datetime(2022,1,11)

# data = web.DataReader('SPY', 'google', start, end)
data = yf.download('SPY', start= start)
```

```python
data
```

```python
for i in range(len(close_df.columns)):
  ticker = close_df.iloc[0,i]
  print(ticker)
  new_df = yf.download(ticker, start=start)
  adj_close = new_df['Adj Close']
  adj_close.name = ticker
  if i==0:
    adj_close_df = adj_close
  else:
    adj_close_df = pd.concat([adj_close_df, adj_close], axis=1)

adj_close_df

# close_df
# df_etf = fdr.DataReader('FEDL', '2019')
# df_etf['Close']
```

### ETF log return df

```python
log_return_df = np.log(etf_close_df/etf_close_df.shift(1))
log_return_df = log_return_df.iloc[1:-7, :]
```
